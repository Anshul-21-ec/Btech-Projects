```python
df_new=df.drop(columns=['GENDER','AGE', 'SMOKING', 'SHORTNESS OF BREATH'])
df_new
```

Finding Correlation

```python
cn=df_new.corr()
cn
```

```python
#Correlation 
cmap=sns.diverging_palette(260,-10,s=50, l=75, n=6,
as_cmap=True)
plt.subplots(figsize=(18,18))
sns.heatmap(cn,cmap=cmap,annot=True, square=True)
plt.show()
```

The correlation matrix shows that ANXIETY and YELLOW_FINGERS are correlated . So, lets create a new feature combining them.

```python
df_new['ANXYELFIN']=df_new['ANXIETY']*df_new['YELLOW_FINGERS']
df_new
```

```python
#Splitting independent and dependent variables
x = df_new.drop('LUNG_CANCER', axis = 1)
y = df_new['LUNG_CANCER']
```

```python
x
```

```python
y
```

Splitting data for training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=32)
```

Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=32)
model.fit(x_train, y_train)
```

Predicting result using testing data

```python
y_pred= model.predict(x_test)
y_pred
```

```python
y_pred_proba = model.predict_proba(x_test)
```

Model accuracy

```python
from sklearn.metrics import  accuracy_score, f1_score,recall_score,precision_score
```

```python
print("Accuracy score is ",round(accuracy_score(y_test, y_pred),2))
```

```python
print("F1 score is ",round(f1_score(y_test, y_pred),2))
```

```python
print("Recall Score is ",round(recall_score(y_test, y_pred),2))
```

```python
print("Precision Score is ",round(precision_score(y_test, y_pred),2))
```

Using RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
```

Predicting result using testing data

```python
y_pred= model2.predict(x_test)
y_pred
```

```python
y_pred_proba = model2.predict_proba(x_test)
```

Model accuracy

```python
print("Accuracy score is ",round(accuracy_score(y_test, y_pred),2))
```

```python
print("F1 score is ",round(f1_score(y_test, y_pred),2))
```

```python
print("Recall Score is ",round(recall_score(y_test, y_pred),2))
```

```python
print("Precision Score is ",round(precision_score(y_test, y_pred),2))
```

ACCURACY AND LOSS GRAPH

```python
from sklearn.metrics import accuracy_score, log_loss
```

```python
accuracy = accuracy_score(y_test, y_pred)
loss = log_loss(y_test, y_pred)

print("Accuracy:", accuracy)
print("Log Loss:", loss)
```

```python
plt.figure(figsize=(10, 5))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(np.arange(len(y_test)), y_test, label='Actual', color='blue')
plt.plot(np.arange(len(y_test)), y_pred, label='Predicted', linestyle='--', color='orange')
plt.title('Actual vs Predicted')
plt.xlabel('Samples')
plt.ylabel('Class')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(np.arange(len(y_test)), -np.log(y_pred_proba[:, 1]), label='Log Loss', color='green')
plt.title('Log Loss')
plt.xlabel('Samples')
plt.ylabel('Log Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
